# Is number happy prime number

```python
def prime(n):
    for i in range(2,n):
        if n % i == 0:
            return  False
        else:
            if n % i == 0:
                break
    return True
# prime(4)
```

```python
def happy(n):
    while n != 1 and n != 4:
        n = sum(int(digit) ** 2 for digit in str(n))
    return n == 1

# happy(7)
```

```python
def is_happy_prime(n):
    if prime(n) and happy(n):
        print('It is happy prime number')
    else:
        print('It is not happy prime number')

is_happy_prime(7)
```

# Strong Number


```python
def strong(n):
    sum = 0
    temp = n
    while n > 0:
        rem = n % 10

        fact = 1
        for i in range(1,rem+1):
            fact *= i

        sum += fact
        n //= 10
    return sum == temp
```

```python
if strong(14):
    print('strong number')
else:
    print('not strong number')
```
